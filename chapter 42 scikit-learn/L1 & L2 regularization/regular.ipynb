```python
df = df.drop(['Address','Date','Postcode','YearBuilt','Lattitude','Longtitude','CouncilArea',], axis=1)
df
```

```python
df.isna().sum()
```

```python
df['Car'] = df['Car'].fillna(0)
df['BuildingArea'] = df['BuildingArea'].fillna(df['BuildingArea'].dropna().mean())
df
```

```python
df[df['Landsize']<df['BuildingArea']]
```

```python
df = df[df['Landsize'] > df['BuildingArea']]
df
```

```python
df.isna().sum()
```

```python
df = pd.get_dummies(df, drop_first=True).astype('float')
df
```

```python
df.columns
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop('Price', axis=1), df.Price, test_size=0.3)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train, y_train)
```

```python
model_L1 = Lasso(alpha=70, tol=0.1, max_iter=100)
model_L1.fit(x_train,y_train)
```

```python
model_L1.score(x_test,y_test)
```

```python
model_L1.score(x_train,y_train)
```

```python
model_L2 = Ridge(alpha=50, max_iter=100, tol=0.1)
model_L2.fit(x_train, y_train)
```

```python
model_L2.score(x_test, y_test)
```

```python
model_L2.score(x_train, y_train)
```
