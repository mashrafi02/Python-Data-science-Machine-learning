```python
import pandas as pd
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, f1_score, precision_score
```

```python
train_df = pd.read_csv('train.csv')
```

```python
label_encoder = LabelEncoder()
train_df['Gender'] = label_encoder.fit_transform(train_df['Gender'])
train_df['Vehicle_Age'] = label_encoder.fit_transform(train_df['Vehicle_Age'])
train_df['Vehicle_Damage'] = label_encoder.fit_transform(train_df['Vehicle_Damage'])
```

```python

train_df.to_csv("numerical_train_df.csv")
```

```python
# Load the dataset inot chunks
chunk_size = 1000000
chunks = pd.read_csv("numerical_train_df.csv", chunksize=chunk_size)
```

```python
# Initialize the classifier
sgd_cls = SGDClassifier(random_state=42)
scaler = StandardScaler()
```

```python
# train on each chunks
for chunk in chunks:
    x_chunk = chunk.drop(["id","Response"], axis = 1)
    y_chunk = chunk["Response"]


    x_chunk = scaler.fit_transform(x_chunk)
    sgd_cls.partial_fit(x_chunk, y_chunk,  classes=[0,1])


```

```python
# Evaluate the model on a test data set
test_df = pd.read_csv('test.csv')
test_df['Gender'] = label_encoder.fit_transform(test_df['Gender'])
test_df['Vehicle_Age'] = label_encoder.fit_transform(test_df['Vehicle_Age'])
test_df['Vehicle_Damage'] = label_encoder.fit_transform(test_df['Vehicle_Damage'])
```

```python
test_df.to_csv("numerical_test.csv")
```

```python
test_df_numeric = pd.read_csv("numerical_train_df.csv")
```

```python
x_test = test_df_numeric.drop(["id","Response"], axis=1)
y_test = test_df_numeric["Response"]
```

```python
x_test = scaler.transform(x_test)
```

```python
y_pred = sgd_cls.predict(x_test)
```

```python
print(y_pred[:50])
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```
