```python
df_uniqe_abilities_values = df["SpecialAbilities"].unique()
df_uniqe_weakness_values = df["Weaknesses"].unique()
print(df_uniqe_abilities_values,df_uniqe_weakness_values)
```

```python
replace_abilities = {"Telekinesis":2,"Invisibility":3,"Super Strength":5,"Flight":2}
replace_weakness = {"Kryptonite":-5,"Magic":-3,"Wooden Stake":-1,"Silver":-2}

df["Abilities_level"] = df["SpecialAbilities"].replace(replace_abilities)
df["Weaknesses_level"] = df["Weaknesses"].replace(replace_weakness)

df.head()

```

```python
df_group_by_character = df.groupby("Character")[["Strength","Speed","Intelligence","Abilities_level","Weaknesses_level","BattleOutcome"]].mean()
df_group_by_character
```

```python
X = df_group_by_character[["Strength","Speed","Intelligence","Abilities_level","Weaknesses_level"]]
Y = df_group_by_character["BattleOutcome"]

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
Y_test
```

```python
scale = StandardScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.transform(X_test)


model = Lasso(alpha=0.1)
model.fit(X_train,Y_train)

Y_pred = model.predict(X_test)

Y_pred
```
