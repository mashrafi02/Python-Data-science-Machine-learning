```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
datas = load_iris()
dir(datas)
```

```python
datas.target_names
```

```python
df = pd.DataFrame(datas.data, columns=datas.feature_names)
df["target"] = datas.target
df["target_names"] = df["target"].apply(lambda x: datas.target_names[x])
```

```python
df
```

```python
df.isna().sum()
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop(["target","target_names"], axis=1), df.target, test_size=0.2, random_state=42)
```

```python
model = RandomForestClassifier() # you can increase the trees with n_estimators attribute
model.fit(x_train, y_train)
```

```python
y_predicted = model.predict(x_test)
y_predicted
```

```python
y_test.values
```

```python
model.score(x_test, y_test)
```
