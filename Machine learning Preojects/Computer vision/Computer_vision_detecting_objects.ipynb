```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
from ultralytics import YOLO
```

```python
# Load the yolov8n model
model = YOLO('yolov8n.pt')
results = model.predict("E:\Gentle Hero\python_course\Machine learning Preojects\Computer vision\web-design-2906159_1280 (1).jpg")
result = results[0]
print(len(result.boxes))
```

```python
for box in result.boxes:
    label = result.names[box[0].cls.item()]
    cords = [round(x) for x in box.xyxy[0].tolist()]
    prob = box.conf[0].item()
    print('object type:', label)
    print('cordinates:', cords)
    print('probability:', prob)
```
